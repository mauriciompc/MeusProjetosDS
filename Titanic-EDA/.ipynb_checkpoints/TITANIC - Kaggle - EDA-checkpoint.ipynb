# Análise Exploratória de Dados do Titanic

### Neste projeto irei realizar uma análise exploratória dos dados do Titanic que podem ser baixados diretamente do [Kaggle](https://www.kaggle.com/competitions/titanic/data). Este dataset está dentro das competições de Machine Learning do site, mas irei utilizar apenas para fazer EDA.

<img src="https://upload.wikimedia.org/wikipedia/commons/6/6e/St%C3%B6wer_Titanic.jpg">

```python
# importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# evitando as mensagens de erro
%matplotlib inline
warnings.filterwarnings ('ignore')
```

```python
# abrindo  e concatenando os arquivos de treino e teste que foram baixados do Kaggle
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')
titanic_df = pd.concat([df1, df2], ignore_index = True)
titanic_df.head()
```

## Descrição das variáveis:

Variável | Descrição | Key
:--- | :--- | :---
PassengerId | Id do passageiro
Survived | Se sobreviveu ou não ao naufrágio | 0 = não, 1 = sim
Pclass | Classe em que viajou (1ª, 2ª ou 3ª Classe) | 1 = 1ª, 2 = 2ª, 3 = 3ª
Name | Nome do Passageiro
Sex | Masculino ou Feminino
Age | Idade
SibSp| Nº de irmãos/cônjuges a bordo do Titanic
Parch | Nº de pais/filhos a bordo do Titanic
Ticket | Número do ticket de embarque
Fare | Tarifa de passageiro
Cabin | Número da Cabine
Embarked | Cidade de Embarque | C = Cherbourg, Q = Queenstown, S = Southampton

```python
# verificando o número de linhas e colunas
# titanic_df.shape
print(f'Temos {titanic_df.shape[0]} linhas e {titanic_df.shape[1]} colunas')
```

```python
# verificando os tipos de dados
titanic_df.info()
```

```python
# verificando a distribuição de sexo dentro do navio
# usaremos a paleta de cor 'Blues' para os gráficos 
sns.countplot(y='Sex', data = titanic_df, palette = 'Blues')
```

```python
# melhorando a visualização de classes x sexo
sns.countplot(x = 'Pclass', data = titanic_df, hue = 'Sex', palette = 'Blues')
```

```python
# criando uma função que vai me dizer se é male, Female ou Child
# para isso vamos definir que quem tiver idade menor que 16 anos será considerado 'child'
def male_female_child(passenger):
    age, sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex

# criando nova coluna com a nova função
titanic_df['person'] = titanic_df[['Age', 'Sex']].apply(male_female_child, axis = 1)
titanic_df.head(10)
```

```python
sns.countplot(x = 'Pclass', data  = titanic_df, hue = 'person', palette = 'Blues')
```

Temos poucas crianças que viajaram na 1ª classe. A grande maioria dos homens estavam viajando na 3ª classe

```python
# verificando a distribuição de idade
titanic_df['Age'].hist(bins = 70)
```

```python
# média de idade dos passageiros
# usando a função round para mostrar o resultado com até 2 casas decimais
idademedia = round(titanic_df['Age'].mean(), 2)
print(f'A média de idade é de: {idademedia} anos')
```

```python
# verificando quantas pessoas temos na nova coluna criada
titanic_df['person'].value_counts()
```

```python
# Porcentagem
titanic_df['person'].value_counts() / titanic_df.shape[0]*100
```

```python
# Verificando a densidade de acordo com Age separado por Sexo
fig = sns.FacetGrid(titanic_df, hue = 'Sex', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0, oldest))
fig.add_legend()
```

```python
# Verificando a densidade de acordo com Age separado por person
fig = sns.FacetGrid(titanic_df, hue = 'person', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0, oldest))
fig.add_legend()
```

```python
# Verificando a densidade de acordo com Age separado por Pclass
fig = sns.FacetGrid(titanic_df, hue = 'Pclass', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0, oldest))
fig.add_legend()
```

```python
# verificando o local de embraque e qual classe dos passageiros
sns.countplot(x = 'Embarked', data = titanic_df, hue = 'Pclass', palette = 'Blues')
```

A grande maioria dos passageiros embarcaram na cidade de Southampton

```python
# Quem estava viajando com a família e quem estaa viajando sozinho?
# Temos a coluna SibSp e Parch que diz se a pessoa estava viajando com a família ou não.

titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
sns.countplot(x = 'Alone', data = titanic_df, palette = 'Blues')
```

```python
sns.countplot(x = 'Alone', data = titanic_df, hue = 'Pclass', palette = 'Blues')
```

A maioria dos passageiros estavam viajando sozinhos, principalmente os homens. Provavelmente para tentar uma nova vida na América, quem sabe?

```python
# Quantidade de quem sobreviveu e quem naufragou?
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no', 1:'yes'})
sns.countplot(x = 'Survivor', data = titanic_df, palette = 'Blues')
```

```python
sns.catplot(kind="point", x = 'Pclass', y = 'Survived', hue ='person', data = titanic_df, palette = 'Blues', ci=None)
```

```python
sns.countplot(x = 'Survivor', hue ='person', data = titanic_df, palette = 'Blues')
```

```python
sns.countplot(x = 'Survivor', hue ='Pclass', data = titanic_df, palette = 'Blues')
```

```python
# quantos sobreviveram ou não. Temos muitos NaN nesta variável.
titanic_df['Survivor'].value_counts()
```

Como já é bem conhecido, poucos sobreviveram ao naufrágio até mesmo porque não havia botes salva vidas para todos já que "nada poderia afundar este navio"! Poucos homens sobreviveram até porque a prioridade dos botes eram de mulheres e crianças, a menos que vc estivesse na primeira classe ou, num surto de seu instinto de sobrevivência, conseguiu entrar em um bote. Mulheres e crianças tiveram maior porcentagem de conseguir sobreviver

Uma teoria recente (nem tanto) leve em consideração que houve uma combustão espontânea no setor de carvoaria, que segundo relatos da época teve início ainda no porto de embarque, e que isto pode ter enfraquecido o casco facilitando a rachadura e consequente inundação quando no choque com o *iceberg*.
