```python
# importar o arquivo listings.csv de Munich para um DataFrame direto do site
# a data dos dados disponibilizado √© de 24 de dez de 2021. 
get_url = "http://data.insideairbnb.com/germany/bv/munich/2021-12-24/visualisations/listings.csv"

df = pd.read_csv(get_url)

# Mostrando todas as colunas (caso n√£o esteja aparecendo).
pd.set_option("display.max.columns", None)
# Mostrando apenas duas casas decimais nas colunas que cont√©m valores.
pd.set_option("display.precision", 2)

df.head()
```

### **Q1. Quantos atributos (vari√°veis) e quantas entradas o nosso conjunto de dados possui? Quais os tipos das vari√°veis?**

```python
# verificando a quantidade de linhas e colunas do dataframe
print(f"Entradas:\t {df.shape[0]}")
print(f"Vari√°veis:\t {df.shape[1]}\n")

# verificando os tipos de dados nas vari√°veis
display(df.dtypes)
```

Podemos observar que temos 18 vari√°veis e 4.995 entradas no dataframe, ou seja, temos dispon√≠veis 4.995 im√≥veis ou quartos para aluguel nesta cidade pelo AirBnb.

Algumas vari√°veis possuem valores nulos e iremos verificar mais adiante se devemos exclu√≠-las do dataframe.

### **Q2. Qual a porcentagem de valores ausentes no *dataset*?**

```python
# ordenar em ordem decrescente as vari√°veis por seus valores ausentes
(df.isnull().sum()).sort_values(ascending=False)
```

Verificamos que as vari√°veis *license* e *neighbourhood_group* n√£o possuem nenhum dado, logo elas podem ser exclu√≠das. Na verdade os dados dessas vari√°veis est√£o em outro dataset dispon√≠vel no site, mas n√£o ser√£o necess√°rias para as an√°lises que irei realizar.

As vari√°veis de *review* (*reviews_per_month e last_review*) tamb√©m possuem muitos dados ausentes pois s√£o campos que dependem da boa vontade de quem alugou dar um feedback.

### **Q3. Qual o tipo de distribui√ß√£o das vari√°veis?** 

```python
# plotar o histograma das vari√°veis num√©ricas
df.hist(bins=15, figsize=(12,8));
```

### **Q4. Qual a m√©dia dos pre√ßos de aluguel?**

```python
# ver a m√©dia da coluna `price`
df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe()
```

Em m√©dia, o pre√ßo do aluguel por noite √© de 127,86 apesar do desvio padr√£o ser
de 374.15. S√≥ neste an√°lise b√°sica j√° vi que dava para ter economizado um bom dinheiro na estadia. Mas ainda preciso ver a quest√£o do bairro pois o deslocamento era algo muito relevante. 


Os dados de algumas vari√°veis tem muitos _outliers_, vamos ent√£o fazer um novo dataset filtrando os dados para uma melhor visualiza√ß√£o dos mesmos e excluindo algumas vari√°veis que n√£o s√£o necess√°rias.

```python
# remover os *outliers* em um novo DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 1500].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 30].index, axis=0, inplace=True)

# remover `neighbourhood_group`, pois est√° vazio
df_clean.drop('neighbourhood_group', axis=1, inplace=True)
df_clean.drop('license', axis=1, inplace=True)
```

```python
# plotar o histograma para as vari√°veis num√©ricas
df_clean.hist(bins=15, figsize=(12,8));
```

### **Q4. Qual a correla√ß√£o existente entre as vari√°veis**

```python
# criar uma matriz de correla√ß√£o
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

# mostrar a matriz de correla√ß√£o
display(corr)
```

```python
# plotar um heatmap a partir das correla√ß√µes
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

Podemos observar que n√£o h√° correla√ß√£o (-0.14) entre o pre√ßo e o m√≠nimo de noites de aluguel. E sendo uma correla√ß√£o negativa √© claro que o pre√ßo diminui quando aumenta o n√∫mero de noites de aluguel e vice-versa.

### **Q5. Qual o tipo de im√≥vel mais alugado no Airbnb?**

```python
# mostrar a quantidade de cada tipo de im√≥vel dispon√≠vel
df_clean.room_type.value_counts()
```

```python
# mostrar a porcentagem de cada tipo de im√≥vel dispon√≠vel
df_clean.room_type.value_counts() / df_clean.shape[0]*100
```

A maioria dos locais para aluguel no AirBnb √© do local inteiro (59%) o que √© bom pois √© o tipo de im√≥vel que buscar√≠amos alugar nesta viagem.

### **Q6. Qual a localidade mais cara do dataset?**



```python
# ver pre√ßos por bairros, na m√©dia
df_clean.groupby(['neighbourhood']).price.mean().sort_values(ascending=False)[:15]
```

```python
# ver quantidade de im√≥veis no Altstadt-Lehel
print(f"N√∫mero de im√≥veis em Altstadt-Lehel: {df_clean[df_clean.neighbourhood == 'Altstadt-Lehel'].shape[0]}")

# ver as 3 entradas do Altstadt-Lehel
df_clean[df_clean.neighbourhood == "Altstadt-Lehel"].head(3)
```

```python
# ver quantidade de im√≥veis no Maxvorstadt 
print(f"N√∫mero de im√≥veis em Maxvorstadt: {df_clean[df_clean.neighbourhood == 'Maxvorstadt'].shape[0]}")

# ver as 3 entradas do Maxvorstadt 
df_clean[df_clean.neighbourhood == "Maxvorstadt"].head(3)
```

Temos mais op√ß√µes de aluguel no bairro Maxvorstadt

```python
# plotar os im√≥veis pela latitude-longitude
df_clean.plot(kind="scatter", x='longitude', y='latitude', alpha=0.4, c=df_clean['price'], s=8,
              cmap=plt.get_cmap('jet'), figsize=(10,6));
```

### **Q7. Qual √© a m√©dia do m√≠nimo de noites para aluguel (minimum_nights)?**

```python
# verificar a m√©dia de quantidade de noites de aluguel da coluna `minimum_nights``
df_clean.groupby(['neighbourhood']).minimum_nights.mean().sort_values(ascending=False)
```

A inten√ß√£o √© ficar mais pr√≥ximo do centro da cidade, ent√£o optei por verificar im√≥veis nos bairros de Maxvorstadt e Altstadt-Lehel. Vamos ent√£o verificar a quantidade m√≠nima e m√°xima de dias de aluguel e o pre√ßo m√©dio dos alugu√©is.

```python
# m√≠nimo e m√°ximo de dias de aluguel no bairro Maxvorstadt
#df_clean.loc[df_clean["neighbourhood"] == "Maxvorstadt", "minimum_nights"].agg(("min", "max"))
print(f"Quantidade m√≠nima de dias de aluguel: {df_clean.loc[df_clean['neighbourhood'] == 'Maxvorstadt', 'minimum_nights'].min()}")
print(f"Quantidade m√°xima de dias de aluguel: {df_clean.loc[df_clean['neighbourhood'] == 'Maxvorstadt', 'minimum_nights'].max()}")
print(f"Quantidade m√©dia de dias de aluguel: {df_clean.loc[df_clean['neighbourhood'] == 'Maxvorstadt', 'minimum_nights'].mean():.2f}")
```

```python
# Pre√ßo do aluguel no bairro Maxvorstadt
#df_clean.loc[df_clean["neighbourhood"] == "Maxvorstadt", "price"].agg(("min", "max"))
print(f"Pre√ßo m√≠nimo em Maxvorstadt: {df_clean.loc[df_clean['neighbourhood'] == 'Maxvorstadt', 'price'].min()}")
print(f"Pre√ßo m√°ximo em Maxvorstadt: {df_clean.loc[df_clean['neighbourhood'] == 'Maxvorstadt', 'price'].max()}")
print(f"Pre√ßo m√©dio em Maxvorstadt: {df_clean.loc[df_clean['neighbourhood'] == 'Maxvorstadt', 'price'].mean():.2f}")
```

```python
# m√≠nimo e m√°ximo de dias de aluguel no bairro Altstadt-Lehel
#df_clean.loc[df_clean["neighbourhood"] == "Altstadt-Lehel", "minimum_nights"].agg(("min", "max"))
print(f"Quantidade m√≠nima de dias de aluguel: {df_clean.loc[df_clean['neighbourhood'] == 'Altstadt-Lehel', 'minimum_nights'].min()}")
print(f"Quantidade m√°xima de dias de aluguel: {df_clean.loc[df_clean['neighbourhood'] == 'Altstadt-Lehel', 'minimum_nights'].max()}")
print(f"Quantidade m√©dia de dias de aluguel: {df_clean.loc[df_clean['neighbourhood'] == 'Altstadt-Lehel', 'minimum_nights'].mean():.2f}")
```

```python
# Pre√ßo do aluguel no bairro Altstadt-Lehel
#df_clean.loc[df_clean["neighbourhood"] == "Altstadt-Lehel", "price"].agg(("min", "max"))
print(f"Pre√ßo m√≠nimo em Altstadt-Lehel: {df_clean.loc[df_clean['neighbourhood'] == 'Altstadt-Lehel', 'price'].min()}")
print(f"Pre√ßo m√°ximo em Altstadt-Lehel: {df_clean.loc[df_clean['neighbourhood'] == 'Altstadt-Lehel', 'price'].max()}")
print(f"Pre√ßo m√©dio em Altstadt-Lehel: {df_clean.loc[df_clean['neighbourhood'] == 'Altstadt-Lehel', 'price'].mean():.2f}")
```
